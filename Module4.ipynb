```python
def myplot(players, param):
    
    for player in players:
        plt.plot(param[Pdict[player]], marker='s', label=player)

    plt.legend(loc="upper left", bbox_to_anchor=(1,1))

    plt.xticks(list(range(0,10)), Seasons, rotation='vertical')
    plt.show()
```

```python
def myplot(param, players = sorted(Players)):
    import itertools
    marker = itertools.cycle(('s', 'o', '^', 'D'))
    
    for player in players:
        plt.plot(param[Pdict[player]], marker=next(marker),ms=12, label=player)

    plt.legend(loc="upper left", bbox_to_anchor=(1,1))

    plt.xticks(list(range(0,10)), Seasons, rotation='vertical')
    plt.show()
```
