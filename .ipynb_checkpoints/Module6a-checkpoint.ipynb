```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('P4-Section6-Homework-Dataset.csv', encoding= 'unicode_escape')
```

```python
df
```

```python
filter = df['Genre'].isin(['action', 'comedy', 'adventure', 'animation', 'drama'])
studios = df['Studio'].isin(['Buena Vista Studios', 'Sony', 'Universal', 'WB', 'Paramount Pictures', 'Fox'])

sns.boxplot(data=df[studios & filter] , x='Genre', y='Gross % US')
sns.scatterplot(data=df[studios & filter], x='Genre', y='Gross % US', hue='Studio')
plt.legend(bbox_to_anchor=(1,1), loc='upper left')
```
